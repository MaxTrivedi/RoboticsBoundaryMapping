```python
def create_interpolated_dataframes(df, left_cols, right_cols, theta_cols):
    dataframes = []

    for left, right, theta in zip(left_cols, right_cols, theta_cols):
        max_theta = df[theta].max()
        min_theta = df[theta].min()

        new_df = pd.DataFrame({'theta': np.arange(max_theta, min_theta - 0.01, -0.01)})
        #new_df[left] = np.nan
        #new_df[right] = np.nan

        #first_left = df[left].iloc[0]
        #last_left = df[left].iloc[-1]
        #first_right = df[right].iloc[0]
        #last_right = df[right].iloc[-1]
        #print(new_df)
        #print(df)
        #try:
        new_df = new_df.merge(df[[left, right, theta]], how='left', left_on='theta', right_on=theta)
        #except:
        print(new_df)


        new_df[left] = new_df[left].interpolate(method='linear')
        new_df[right] = new_df[right].interpolate(method='linear')

        new_df = new_df.drop(columns=[theta])
        print(new_df)

        dataframes.append(new_df)


    return dataframes
'''
        new_df.at[0, left] = first_left
        print(first_left)
        new_df.at[len(new_df) - 1, left] = last_left
        print(last_left)
        new_df.at[0, right] = first_right
        new_df.at[len(new_df) - 1, right] = last_right

        # interpolation
        new_df[left] = new_df[left].interpolate()
        new_df[right] = new_df[right].interpolate()

        '''


left_columns = ['Left_1', 'Left_2', 'Left_3', 'Left_4', 'Left_5']
right_columns = ['Right_1', 'Right_2', 'Right_3', 'Right_4', 'Right_5']
theta_columns = ['theta_1', 'theta_2', 'theta_3', 'theta_4', 'theta_5']

interpolated_dfs = create_interpolated_dataframes(df, left_columns, right_columns, theta_columns)



```
